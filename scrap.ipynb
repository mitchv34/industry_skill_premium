```python
d1 = pd.read_csv("./data/raw/inv_ip.csv", sep = ";")
d1.SeriesCode = d1.SeriesCode.apply(lambda x: x[3:7])
d1.loc[1:, ["SeriesCode", "LineDescription"]]
```
